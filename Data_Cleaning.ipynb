```python
# The first column contains string ids from which you need to extract only the numerical part and convert them to integers.
for i in range(len(df['Unique ID'])):
  df['Unique ID'][i] = df['Unique ID'][i][3:]
```

```python
# Next we will use those extracted numbers and set them as the primary index. We can achieve this using the set_index function.
df.set_index('Unique ID',inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Apply label encoding on the columns converting them to integer so that it can be used to train machine learning models. You can use sklearn for this or explore other methods.

labelencoder = LabelEncoder()
df['Address_int'] = labelencoder.fit_transform(df['Address'])
df['Suburb_int'] = labelencoder.fit_transform(df['Suburb'])
df['SellerG_int'] = labelencoder.fit_transform(df['SellerG'])
df['Regionname_int'] = labelencoder.fit_transform(df['Regionname'])
df['Method_int'] = labelencoder.fit_transform(df['Method'])
df['CouncilArea_int'] = labelencoder.fit_transform(df['CouncilArea']);
```

```python
# df.fillna('Unknown',inplace = True)
df.isnull().sum()
```

```python
# filling null values of Car with the mode of the Car column
val = df['Car'].mode()[0]
df['Car'] = df['Car'].fillna(val)
```

```python
# filling null values of BuildingArea with the mean of the BuildingArea column
val = df['BuildingArea'].mean()
df['BuildingArea'] = df['BuildingArea'].fillna(val)
```

```python
# filling null values of YearBuilt with the mode of the YearBuilt column
val = df['YearBuilt'].mode()[0]
df['YearBuilt'] = df['YearBuilt'].fillna(val)
```

```python
# filling null values of CouncilArea with the mode of the CouncilArea column
val = df['CouncilArea'].mode()[0]
df['CouncilArea'] = df['CouncilArea'].fillna(val)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
arr = ['Distance','Car','BuildingArea','YearBuilt','Lattitude','Longtitude']
for i in arr:
  df[i] = df[i].astype('int')
```

```python
room_mean = df['Rooms'].mean();
room_mode = df['Rooms'].mode()[0];
room_std = df['Rooms'].std()
room_median = df['Rooms'].median()

price_mean = df['Price'].mean();
price_mode = df['Price'].mode()[0];
price_std = df['Price'].std()
price_median = df['Price'].median()

print("Mean of Rooms is",room_mean,sep = ' ')
print("Mode of Rooms is",room_mode,sep = ' ')
print("Standard Deviation of Rooms is",room_std,sep = ' ')
print("Median of Rooms is",room_median,sep = ' ')
print()
print("Mean of Price is",price_mean,sep = ' ')
print("Mode of Price is",price_mode,sep = ' ')
print("Standard Deviation of Price is",price_std,sep = ' ')
print("Median of Price is",price_median,sep = ' ')
```
